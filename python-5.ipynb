```python
df1=pd.DataFrame({"sub":["FSD","Python","DM"],"fac":["psp","vha","mss"],"id":[1,2,4]})
df2=pd.DataFrame({"id":[1,2,3],"marks":[9,10,8]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="id",how="inner")
```

```python
pd.merge(df1,df2,on="id")
```

```python
pd.merge(df1,df2,on="id",how="outer")
```

```python
df1.merge(df2,on="id")
```

```python
df1.merge(df2,on="id",how="left")#only for df1
```

```python
df1.merge(df2,on="id",how="right")
```

```python
df1.merge(df2,how="cross")
# all possible combination
```

```python
df1=pd.DataFrame({"sub":["FSD","Python","DM"],"fac":["psp","vha","mss"],"id_x":[1,2,4]})
df2=pd.DataFrame({"id_v":[1,2,3],"marks":[9,10,8]})
```

```python
df1.merge(df2,left_on="id_x",right_on="id_v",right_index=True,left_index=True) #col name diff, but data same
```
